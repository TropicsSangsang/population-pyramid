## 남자 대이터 정의

```python
import pandas as pd
file_name = '202306_202306_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_m
```

```python
import pandas as pd
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_m.head(3) # 3개만 가져옴
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int)
# 0번째 데이터(현재 str)를 ','를 없앤후 int형으로 바꾼 데이터를 다시 df_m.iloc[0] 에 저장 : 1,188,733 -->1188733(정수형)
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AY:BS')
df_w.head(3)
```

- Column 명에 .1 이런거를 없는 것으로 통일시키기 위해 전국 단위 한번 불러오고, 그것의 컬럼으로 대체

```python
df_n = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_n.head(1)
```

```python
df_n.columns
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_m.columns = df_n.columns # 컬럼명 통일
df_m.columns
```

```python
df_w.columns = df_n.columns # 컬럼명 통일
df_w.columns
```

- 다시 여자 데이터 정의

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
# 0번째 데이터(현재 str)를 ','를 없앤후 int형으로 바꾼 데이터를 다시 df_m.iloc[0] 에 저장 : 1,188,733 -->1188733(정수형)
df_w.head(3)
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우인 경우
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Apple 인 경우
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.barh(df_m.columns, df_m.iloc[0] // 1000)  # 단위 : 천 명
```

```python
plt.figure(figsize=(10, 7))   # 사이즈를 가로로 좀 키우고...
plt.barh(df_m.columns, df_m.iloc[0] // 1000)  # 단위 : 천 명
```

```python
plt.figure(figsize=(10, 7))   # 사이즈를 가로로 좀 키우고...
plt.barh(df_m.columns, df_m.iloc[0] // 1000)  # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000) # 비슷해서 구분하기가;;--> 이럴때는 남자쪾에 - 를 붙여서...아래쪽으로 GoGo!!
```

```python
plt.figure(figsize=(10, 7))   # 사이즈를 가로로 좀 키우고...
plt.barh(df_m.columns, -df_m.iloc[0] // 1000)  # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
```

```python
plt.figure(figsize=(10, 7))   # 사이즈를 가로로 좀 키우고...
plt.barh(df_m.columns, -df_m.iloc[0] // 1000)  # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2023 대한민국 인구 피라미드')
plt.savefig('2023_인구피라미드.png', dpi=100)
plt.show()
```
